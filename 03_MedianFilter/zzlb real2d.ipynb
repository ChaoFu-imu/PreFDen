```python
import numpy as np
import scipy.io as sio
from scipy.ndimage import median_filter

# ====================== load data ======================
data = sio.loadmat('real2det(1).mat')
Dn=data['real2det']

# ====================== Median Filter Denoising ======================
denoised_data = median_filter(Dn, size=(5,3))

# ====================== sava data======================
save_dict = {
    'D_denoised': denoised_data,
}

# if 'Dc' in locals() or 'Dc' in globals():
#     save_dict['true_noise'] = Dn - Dc

output_path = 'zzlb 2d real_result 5，3 .mat'
sio.savemat(output_path, save_dict)
print(f'【Median Filter Denoised result saved to：{output_path}】')
```
