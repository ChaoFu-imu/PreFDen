```python
import numpy as np
import scipy.io as sio
from scipy.ndimage import median_filter
import h5py

# ================== load data ==================
with h5py.File('seis3dsyn.mat', 'r') as f:
    # Read and transpose data to align with [nz, nx, ny] dimensions
    Dc = np.array(f['DataClean']).T  # Clean data
    Dn = np.array(f['DataNoisy']).T  # Noisy data

print(f"Clean data shape: {Dc.shape}, Noisy data shape: {Dn.shape}")

# ================== 3D Median Filter Denoising ==================
denoised_data = median_filter(Dn, size=(3, 3, 3))

# ================== Compute SNR (fully adapted for 3D)==================
def compute_snr(clean, noisy):
    noise = noisy - clean
    return 10 * np.log10(np.sum(clean**2) / np.sum(noise**2))

snr_before = compute_snr(Dc, Dn)
snr_after = compute_snr(Dc, denoised_data)
print(f"SNR before denoising: {snr_before:.2f} dB")
print(f"SNR after denoising: {snr_after:.2f} dB")
print(f"SNR improvement: {snr_after - snr_before:.2f} dB")

# ================== Save data==================
save_dict = {
    'D_denoised': denoised_data
}

output_path = 'zzlb_3d_syn_result 3 3 3.mat'
sio.savemat(output_path, save_dict, do_compression=True)
print(f"[3D Median Filter] Denoised result saved to: {output_path}】")
print(f"Variable name: D_denoised, data shape: {denoised_data.shape}")
```
