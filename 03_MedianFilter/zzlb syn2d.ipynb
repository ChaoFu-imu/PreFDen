```python
import numpy as np
import scipy.io as sio
from scipy.ndimage import median_filter
import matplotlib.pyplot as plt

# ====================== load data ======================
data = sio.loadmat('seismic_snr_-1dB.mat')
# dt = data['Nt']
# dx = data['Nx']
Dc = data['DataClean']
Dn = data['DataNoisy']

# ====================== Median Filter Denoising ======================
denoised_data = median_filter(Dn, size=(5,3))

# ====================== Compute SNR ======================
def compute_snr(clean, noisy):
    noise = noisy - clean
    return 10 * np.log10(np.sum(clean**2) / np.sum(noise**2))

snr_before = compute_snr(Dc, Dn)
snr_after = compute_snr(Dc, denoised_data)
print(f"SNR before denoising: {snr_before:.2f} dB")
print(f"SNR after denoising: {snr_after:.2f} dB")
print(f"SNR improvement:  {snr_after - snr_before:.2f} dB")

# ====================== sava data ======================
save_dict = {
    'D_denoised': denoised_data,
}

if 'Dc' in locals() or 'Dc' in globals():
    save_dict['true_noise'] = Dn - Dc

output_path = 'zzlb 2d syn_result5.mat'
sio.savemat(output_path, save_dict)
print(f'【Median Filter Denoised result saved to：{output_path}】')
```
